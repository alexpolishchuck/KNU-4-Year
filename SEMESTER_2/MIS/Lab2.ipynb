1) Вводиться число n. Вивести масив розміру n×n, в
якому по діагоналі йдуть числа від 1 до n, а інші числа дорівнюють 0.

```python
def create_diagonal_matrix(n):
    start_time = time.time()

    matrix = np.zeros((n, n), dtype=int)
    np.fill_diagonal(matrix, np.arange(1, n + 1))

    end_time = time.time()

    return matrix, end_time - start_time


def create_diagonal_matrix_no_numpy(n):

    start_time = time.time()

    matrix = [[0 for _ in range(n)] for _ in range(n)]
    for i in range(n):
        matrix[i][i] = i + 1

    end_time = time.time()

    return matrix, end_time - start_time

n = 1000

matrix, time_numpy = create_diagonal_matrix(n)

print(f"Час виконання з NumPy: {time_numpy:.6f} секунд")
#print(matrix)

matrix, time_no_numpy = create_diagonal_matrix_no_numpy(n)

print(f"Час виконання без NumPy: {time_no_numpy:.6f} секунд")
#print(matrix)
```

3) Вводяться 4 числа n, m, r, c. Вивести масив розміру
n×m, в якому в кожному рядку з номером r і в кожному стовпчику з
номером c стоять 0, а інші елементи дорівнюють 1

```python
def create_matrix_with_zeros(n, m, r, c):

    start_time = time.time()

    matrix = np.ones((n, m), dtype=int)
    matrix[r, :] = 0
    matrix[:, c] = 0

    end_time = time.time()

    return matrix, end_time - start_time


def create_matrix_with_zeros_no_numpy(n, m, r, c):
    start_time = time.time()

    matrix = [[1 for _ in range(m)] for _ in range(n)]
    for i in range(n):
        matrix[i][c] = 0
    for j in range(m):
        matrix[r][j] = 0

    end_time = time.time()

    return matrix, end_time - start_time


n = 1000
m = 1000
r = 2
c = 1

matrix, time_numpy = create_matrix_with_zeros(n, m, r, c)

print(f"Час виконання з NumPy: {time_numpy:.6f} секунд")
#print(matrix)

matrix, time_no_numpy = create_matrix_with_zeros_no_numpy(n, m, r, c)

print(f"Час виконання без NumPy: {time_no_numpy:.6f} секунд")
#print(matrix)
```

6) З клавіатури вводиться масив. Замінити всі ненульові
елементи на –1.

```python
def replace_non_zero_elements(array):
    start_time = time.time()

    array[array != 0] = -1

    end_time = time.time()

    return array, end_time - start_time

def replace_non_zero_elements_no_numpy(array):
    start_time = time.time()

    for i in range(len(array)):
        for j in range(len(array[0])):
            if array[i][j] != 0:
                array[i][j] = -1

    end_time = time.time()

    return array, end_time - start_time

array = np.array([[1, 2, 3], [4, 0, 6], [7, 8, 9]])

array, time_numpy = replace_non_zero_elements(array)

print(f"Час виконання з NumPy: {time_numpy:.6f} секунд")
print(array)

array = [[1, 2, 3], [4, 0, 6], [7, 8, 9]]
array, time_no_numpy = replace_non_zero_elements_no_numpy(array)

print(f"Час виконання без NumPy: {time_no_numpy:.6f} секунд")
print(array)
```

8. З клавіатури вводиться масив. Підрахувати в ньому
кількість нульових та ненульових елементів.

```python
def count_zeros_and_nonzeros():
  
    array = list(map(int, input("Введіть елементи масиву через пробіл: ").split()))
    
    count_zeros = 0
    count_nonzeros = 0
    start_time = time.time()
    
    for element in array:
        if element == 0:
              count_zeros += 1
        else:
              count_nonzeros += 1
    end_time = time.time()
    print(f"Час виконання: {end_time - start_time:.6f} секунд")
    
    return count_zeros, count_nonzeros

def count_zeros_and_nonzeros_numpy():
    array = np.array(list(map(int, input("Введіть елементи масиву через пробіл: ").split())))
    
    start_time = time.time()
    
    count_zeros = np.count_nonzero(array == 0)
    count_nonzeros = len(array) - count_zeros
    
    end_time = time.time()
    print(f"Час виконання: {end_time - start_time:.6f} секунд")
    
    return count_zeros, count_nonzeros

count_zeros, count_nonzeros = count_zeros_and_nonzeros()

print(f"Кількість нульових елементів: {count_zeros}")
print(f"Кількість ненульових елементів: {count_nonzeros}")

count_zeros, count_nonzeros = count_zeros_and_nonzeros_numpy()

print(f"Кількість нульових елементів numpy: {count_zeros}")
print(f"Кількість ненульових елементів numpy: {count_nonzeros}")
```

11) Створити масив одиниць розміром n×n та додати до нього «рамку», що утворюється з 0

```python
import numpy as np

def create_array_with_frame(n):
    
    start_time = time.time()
    
    array = np.ones((n, n), dtype=int)
    array[0, :] = 0
    array[-1, :] = 0
    array[:, 0] = 0
    array[:, -1] = 0
    
    end_time = time.time()
    print(f"Час виконання: {end_time - start_time:.20f} секунд")
    
    return array

def create_array_with_frame_no_numpy(n):
    
    start_time = time.time()
    
    array = [[0 for _ in range(n)] for _ in range(n)]
    for i in range(1, n - 1):
        for j in range(1, n - 1):
            array[i][j] = 1
            
    end_time = time.time()
    print(f"Час виконання: {end_time - start_time:.20f} секунд")
    
    return array

n = 100

array = create_array_with_frame_no_numpy(n)
for row in array:
    print(row)

# numpy
array = create_array_with_frame(n)

print(array)

```

16) Заповнити вектор значеннями від 0 до n. Всі значення,
що більші за n/4 та менші за 3n/4 замінити нулями.

```python
def create_vector_with_zeros(n):
    start_time = time.time()

    vector = np.arange(n)
    vector[(vector > n / 4) & (vector < 3 * n / 4)] = 0

    end_time = time.time()

    return vector, end_time - start_time


def create_vector_with_zeros_no_numpy(n):

    start_time = time.time()

    vector = []
    for i in range(n):
        if i > n / 4 and i < 3 * n / 4:
            vector.append(0)
        else:
            vector.append(i)

    end_time = time.time()

    return vector, end_time - start_time

n = 10000

vector, time_numpy = create_vector_with_zeros(n)

print(f"Час виконання з NumPy: {time_numpy:.6f} секунд")
print(vector)

vector, time_no_numpy = create_vector_with_zeros_no_numpy(n)

print(f"Час виконання без NumPy: {time_no_numpy:.6f} секунд")
print(vector)
```

18. Згенерувати вектор з n випадкових чисел в діапазоні
від 0 до 99. Визначити кількість унікальних чисел в послідовності.

```python
def generate_unique_numbers(n):
    start_time = time.time()

    vector = np.random.randint(0, 100, n)
    unique_numbers = np.unique(vector).size

    end_time = time.time()

    return vector, unique_numbers, end_time - start_time

def generate_unique_numbers_no_numpy(n):
    start_time = time.time()

    vector = []
    for _ in range(n):
        vector.append(random.randint(0, 99))
    unique_numbers = len(set(vector))

    end_time = time.time()

    return vector, unique_numbers, end_time - start_time

n = 100000

vector, unique_numbers, time_numpy = generate_unique_numbers(n)

print(f"Кількість унікальних чисел: {unique_numbers}")
print(f"Час виконання з NumPy: {time_numpy:.6f} секунд")

vector, unique_numbers, time_no_numpy = generate_unique_numbers_no_numpy(n)

print(f"Кількість унікальних чисел: {unique_numbers}")
print(f"Час виконання без NumPy: {time_no_numpy:.6f} секунд")
```

19. Згенерувати вектор із n елементів, що рівномірно
розміщені на інтервалі (0,1) – тобто обидва кінці інтервалу не
включені. Значення вивести до 3 знаків після коми.

```python
def generate_random_vector(n):

    start_time = time.time()

    vector = np.random.uniform(0, 1, n)
    vector = np.round(vector, 3)

    end_time = time.time()

    return vector, end_time - start_time

def generate_random_vector_no_numpy(n):
    start_time = time.time()

    vector = []
    for _ in range(n):
        vector.append(round(random.uniform(0, 1), 3))

    end_time = time.time()

    return vector, end_time - start_time


# Приклад використання

n = 100000

vector, time_numpy = generate_random_vector(n)

print(f"Час виконання з NumPy: {time_numpy:.6f} секунд")
print(vector)

vector, time_no_numpy = generate_random_vector_no_numpy(n)

print(f"Час виконання без NumPy: {time_no_numpy:.6f} секунд")
print(vector)
```

```python
def cramer_method(A, b):

    det_A = np.linalg.det(A)
    print(det_A)
    
    print(np.insert(np.delete(A, 0, 1), 0, b, axis=1))
    det_A_x = np.linalg.det(np.insert(np.delete(A, 0, 1), 0, b, axis=1))
    print(det_A_x)
    det_A_y = np.linalg.det(np.insert(np.delete(A, 1, 1), 1, b, axis=1))
    print(det_A_y)
    det_A_z = np.linalg.det(np.insert(np.delete(A, 2, 1), 2, b, axis=1))
    print(det_A_z)
    det_A_f = np.linalg.det(np.insert(np.delete(A, 3, 1), 3, b, axis=1))
    print(det_A_z)

    x = det_A_x / det_A
    y = det_A_y / det_A
    z = det_A_z / det_A
    f = det_A_f / det_A

    return np.array([x, y, z, f])

A = np.array([[1, 2, 3, 4], [2, 1, 2, 3], [3, 2, 1, 2], [4 ,3 ,2 , 1]])
b = np.array([5, 1, 1, -5])

solution_cramer = cramer_method(A, b)

print("Розв'язок за допомогою формул Крамера:", solution_cramer)

solution_matrix_multiplication = np.linalg.solve(A, b)

print("Перевірка за допомогою матричного множення:", solution_matrix_multiplication)

inverse_A = np.linalg.inv(A)

solution_inverse_matrix = np.dot(inverse_A, b)

print("Перевірка за допомогою оберненої матриці:", solution_inverse_matrix)


solution_linalg_solve = np.linalg.solve(A, b)

print("Перевірка за допомогою numpy.linalg.solve():", solution_linalg_solve)

print("Порівняння всіх рішень:", np.allclose(solution_cramer, solution_matrix_multiplication), np.allclose(solution_cramer, solution_inverse_matrix), np.allclose(solution_cramer, solution_linalg_solve))
```

$$
3)\:2(A - B)(A^{2} + B)
$$

A = \begin{bmatrix}5&1&7\\-10&-2&1\\0&1&2\end{bmatrix} B = \begin{bmatrix}2&4&1\\3&1&0\\7&2&1\end{bmatrix}

```python
def matrix_expression(A, B):
    start_time = time.time()

    C = 2 * (A - B) * (np.dot(A, A) + B)

    end_time = time.time()

    return C, end_time - start_time


def matrix_expression_no_numpy(A, B):
    start_time = time.time()

    C = [[0 for _ in range(len(A[0]))] for _ in range(len(A))]

    for i in range(len(A)):
        for j in range(len(A[0])):
            for k in range(len(A[0])):
                C[i][j] += 2 * (A[i][k] - B[i][k]) * (A[k][j] + B[k][j])

    end_time = time.time()

    return C, end_time - start_time

A = np.array([[5, 1, 7], [-10, -2, 1], [0, 1, 2]])
B = np.array([[2, 4, 1], [3, 1, 0], [7, 2, 1]])

C_numpy, time_numpy = matrix_expression(A, B)

print(f"Час виконання з NumPy: {time_numpy:.6f} секунд")
print(C_numpy)

C_no_numpy, time_no_numpy = matrix_expression_no_numpy(A, B)

print(f"Час виконання без NumPy: {time_no_numpy:.6f} секунд")
print(C_no_numpy)
```
